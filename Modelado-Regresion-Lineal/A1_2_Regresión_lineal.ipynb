```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
##from google.colab import files
##uploaded = files.upload()

df = pd.read_csv("A1.2 Felicidad y GDP.csv")
df.head()

df.info()
df.describe()
```

```python
df["log_GDP"] = np.log(df["GDP"])
```

```python
plt.figure()
plt.scatter(df["log_GDP"], df["Felicidad"])
plt.xlabel("log(GDP)")
plt.ylabel("Felicidad")
plt.title("Relación entre Felicidad y log(GDP)")
plt.show()
```

```python
X = sm.add_constant(df["log_GDP"])
y = df["Felicidad"]

modelo_simple = sm.OLS(y, X).fit()
print("Coeficiente log(GDP):", modelo_simple.params["log_GDP"])
print("p-value log(GDP):", modelo_simple.pvalues["log_GDP"])
print("R-squared:", modelo_simple.rsquared)
print("F-statistic:", modelo_simple.fvalue)
```

## Interpretación

El modelo de regresión lineal simple muestra una relación positiva y estadísticamente significativa entre el logaritmo del GDP y el nivel de felicidad. El coeficiente asociado al log(GDP) es positivo y presenta un p-value menor a 0.05, lo que indica que el GDP es un factor relevante en la explicación de la felicidad. Sin embargo el valor de R² ≈ 0.22 sugiere que el GDP explica solo una parte de la variabilidad observada, confirmando que la felicidad es un fenómeno multidimensional.

## Limitaciones y extensión

El análisis realizado mediante regresión lineal simple presenta limitaciones importantes ya que no considera factores sociales, de salud o institucionales que influyen en la felicidad. Para obtener un modelo más realista es necesario extender el análisis a una regresión lineal múltiple incorporando variables adicionales como la esperanza de vida, el apoyo social y la percepción de corrupción.

## Conclusión

En conclusión, el GDP presenta una relación positiva y significativa con el nivel de felicidad de los países, sin embargo no es suficiente para explicar completamente este fenómeno. El uso de modelos de regresión lineal permite identificar relaciones importantes entre variables pero la felicidad requiere un enfoque multidimensional que considere factores económicos y sociales de manera conjunta.
